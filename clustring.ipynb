```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
```

```python
import pandas as pd
df = pd.read_csv("/content/customers.csv")
```

```python
x = df[['AverageSpend','AverageFrequency']]
```

```python
from sklearn.cluster import AgglomerativeClustering
agg_model = AgglomerativeClustering(n_clusters=10)
```

```python
df['cluster'] = agg_model.fit_predict(x)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x['AverageSpend'],x['AverageFrequency'],c=df['cluster'],cmap='rainbow')
plt.xlabel('AverageSpend')
plt.ylabel('AverageFrequency')
plt.title("Agglomerative Clustering Example")
plt.show()
```
